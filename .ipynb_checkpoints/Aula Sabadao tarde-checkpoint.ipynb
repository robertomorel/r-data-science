### 1 - Gerar a partir dos dados, informações de valores máximos, mínimos, quartis, amplitude, variância, em relação à $arvore$ por $tronco$

```r
tronco_por_arvore <- data.frame(arvore=sort(unique(esaligna$arvore)),
                            maximo=with(esaligna, tapply(tronco, arvore, max)),
                            minimos=with(esaligna, tapply(tronco, arvore, min)),
                             quartis=with(esaligna, tapply(tronco, arvore, quantile)),
                             amplitude=with(esaligna, tapply(tronco, arvore, range)),
                             variancia=with(esaligna, tapply(tronco, arvore, var)),
                             media=with(esaligna, tapply(tronco, arvore, mean)),
                             std=with(esaligna, tapply(tronco, arvore, sd))

)
```

```r
tronco_por_arvore
```

```r
summary(esaligna)
```

```r
summary(esaligna[esaligna$arvore == 1,"tronco"])
```

```r
p_media_desvio <- ggplot(tronco_por_arvore, aes(factor(arvore), media))+
                        geom_point(stat="identity",size=5)+
                        geom_errorbar(aes(ymin=media-std, ymax=media+std), with=0.5)+
                        xlab("arvore")
```

```r
p_media_desvio
```

```r
set.seed(123456789)
```

```r
grpMeat <- kmeans(food[,c("RedMeat","WhiteMeat")],centers=4,nstart=10)
```

```r
grpMeat
```

```r
o = order(grpMeat$cluster)
o
```

```r
data.frame(food$Country[o], grpMeat$cluster[o])
```

```r
plot(food$Red, food$White, type="n", xlim=c(3,19), xlab="Carne Vermelha", ylab="Carne branca")
text(x=food$Red, y=food$White, food$Country, col=grpMeat$cluster)
```

```r
teens <- read.csv("https://raw.githubusercontent.com/brenden17/sklearnlab/master/facebook/snsdata.csv")
head(teens,3)
dim(teens)
```

```r
interesses<- teens[5:40]
```

```r
intesesses <- as.data.frame(na.omit(interesses))
```

```r
interesses
```

```r
interesses.stand <- scale(interesses) 
```

```r
head(interesses.stand)
```

```r
interesses.clusters <- kmeans(interesses.stand, 15)
```

```r
interesses.clusters$size
```

```r
interesses.clusters$centers
```

```r
pie(colSums(interesses[interesses.clusters$cluster==10,]),cex=1)
```

```r
k.max <- 100
```

```r
dados <- interesses.stand
```

```r
dados
```

```r
wss <- sapply(1:k.max, 
              function(k){kmeans(dados,k,nstart=5,iter.max=15)$tot.withinss})
```

```r
wss
```

```r
plot(1:k.max, wss,
    type="b", pch=19, frame=FALSE,
    xlab="Número de clusters K",
    ylab="Total within-clusters sum of squares")
```
