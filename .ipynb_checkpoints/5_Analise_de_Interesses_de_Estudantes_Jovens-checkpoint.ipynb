```r
# Busca de base de dados de prática de esportes de jovens por formação
teens <- read.csv("https://raw.githubusercontent.com/brenden17/sklearnlab/master/facebook/snsdata.csv")
head(teens,3)
dim(teens)
```

```r
# Sumarização
summary(teens)
```

```r
# Remover dados faltantes
teens_no_na <- subset(teens, !is.na(teens$age) & !is.na(teens$gender))
teens_no_na
```

```r
# Retira a linha inteira caso algum campo esteja vazio (NA)
teens_clean <- na.omit(teens)
teens_clean
```

```r
# Filtrar datafrma somente por interesse dos jovens
#interesses <- subset(teens[,5:40])
# teens[-1,-4] # Remover as 4 primeiras colunas
interesses <- as.data.frame(teens[5:40])
interesses
```

```r
# Escalar interesses para não considerar peso e atrapalhar a análise
interesses.stand <- scale(interesses)
interesses.stand
```

```r
# Criando clusters para agrupar os interesses comuns
# Gerou 15 grupos diferentes por interesse
interesses.clusters <- kmeans(interesses.stand, 15)

attributes(interesses.clusters)

interesses.clusters$size

interesses.clusters$centers
```

```r
# Criar um gráfico de pizza para ilustras os clusters
# Segmentação de clientes
pie(colSums(interesses[interesses.clusters$cluster==10,]),cex=1)
```

```r
# Valor máximo de testes
k.max <- 50
dados <- interesses.stand
dados
```

```r
# Pegando valores de 1-50 e para cara valor, será aplicado o algorítmo de clustorização
wss <- sapply(1:k.max, 
              function(k){kmeans(dados,k,nstart=5,iter.max=15)$tot.withinss})
wss
```

```r
# Para saber o número de clusters bons para serem usados 
# Quanto mais perto os pontos ficarem um do outro, melhor
plot(1:k.max, wss,
    type="b", pch=19, frame=FALSE,
    xlab="Número de clusters K",
    ylab="Total within-clusters sum of squares")
```
